```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome("E://chromedriver")
```

```python
str2='https://www.vegrecipesofindia.com/recipes/no-onion-no-garlic-recipes/page/'
recipes=[]
for i in range(1,32):
    driver.get(str2+str(i))
    g=driver.find_elements_by_xpath('//a[@href]')
    rec=[]
    for i in g:
        rec.append(i.get_attribute("href"))
    temp1=78
    temp2=111
    rec=list(set(rec))
    finaltest=[]
    for i in rec:
        if('https://www.vegrecipesofindia.com/recipes/' not in i):
            finaltest.append(i)
    recipes.append(finaltest)
```

```python
a='https://www.vegrecipesofindia.com/press-media/'
b='https://www.youtube.com/user/vegrecipesofindia'
c='https://www.vegrecipesofindia.com/terms-and-conditions/'
d='https://www.vegrecipesofindia.com/privacy-policy/'
e='https://www.facebook.com/vegrecipesofindia'
f='https://www.instagram.com/vegrecipesofindia/'
g='https://www.vegrecipesofindia.com/'
h='https://twitter.com/vegrecipeindia'
i='https://www.vegrecipesofindia.com/about/'
j='https://in.pinterest.com/vegrecipesofindia/'
tobedel=[a,b,c,d,e,f,g,h,i,j]
```

```python
finalrec=[]
for i in range(len(recipes)):
    for k in range(len(recipes[i])):
        if(recipes[i][k] not in tobedel):
            finalrec.append(recipes[i][k])
```

```python
len(finalrec)
```

```python
link=[]
title=[]
cuisine=[]
totaltime=[]
ings=[]
```

```python
for i in range(len(finalrec)):
    print(i)
    str1=finalrec[i]
    link.append(finalrec[i])
    x=[]
    driver.get(str1)
    a=driver.find_element_by_class_name( "type-post") 
    d=a.text.split('\n')
    temp1=d.index('Ingredients')
    temp2=d.index('Instructions')
    ingstest=d[temp1:temp2]
    ings.append(ingstest)
    title.append(d[0])
    temp1=d.index('Total Time')
    totaltime.append(d[temp1+1])
    temp1=d.index('Cuisine')
    cuisine.append(d[temp1+1])
```

```python
df=pd.DataFrame({"URL":link,'Dish':title,'Cuisine':cuisine,'Total Time': totaltime,'Ingredients':ings})
```
