```python
def vectorizer(sent,m):
    vec= [] 
    numw = 0
    for w in sent:
        try:
            if numw==0:
                vec = m[w]
            else:
                vec = np.add(vec,m[w])
            numw+=1
        except:
            pass
    
    return np.asarray(vec) /numw
```

```python
dbtest.Style.unique()
dbtest["StyleNum"] = dbtest["Style"].astype('category').cat.codes
dbtest["CuisineNum"] = dbtest["Cuisine"].astype('category').cat.codes
```

```python
dbtest.StyleNum.unique()
```

```python
l = []
for i in range(len(dbtest)):
    l.append(vectorizer(dbtest.iloc[i][4],ingmodel))
```

```python
for i in range(len(l)):
    l[i] = np.append(l[i],[dbtest.iloc[i][3],dbtest.iloc[i]['StyleNum']])
```

```python
X = np.array(l)
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1,100):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,100),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
