```python
for i in range(len(finalpd)):
    day,hour,mins='','',''
    if( 'd' in finalpd.iloc[i][3]):
        d=finalpd.iloc[i][3].index('d')
        day=finalpd.iloc[i][3][:d]
        finalpd.iloc[i][3]=finalpd.iloc[i][3][d+1:]
        day=day.strip()
       
    if( 'hrs' in finalpd.iloc[i][3]):
        h=finalpd.iloc[i][3].index('h')
        hour=finalpd.iloc[i][3][:h]
        finalpd.iloc[i][3]=finalpd.iloc[i][3][h+3:]
        hour=hour.strip()
        
    if( 'hr' in finalpd.iloc[i][3]):
        h=finalpd.iloc[i][3].index('h')
        hour=finalpd.iloc[i][3][:h]
        finalpd.iloc[i][3]=finalpd.iloc[i][3][h+2:]
        hour=hour.strip()
    
    if('m' in finalpd.iloc[i][3]):
        m=finalpd.iloc[i][3].index('m')
        mins=finalpd.iloc[i][3][:m]
        mins=mins.strip()
        
    if(day==''):
        day=0
    else:
        day=int(day)
    if(hour==''):
        hour=0
    else:
        hour=int(hour)
    if(mins==''):
        mins=0
    else:
        mins=int(mins)
    finalpd.iloc[i][3]=(day,hour,mins)
#     print(finalpd.iloc[i][3])
```
