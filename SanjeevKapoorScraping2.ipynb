```python
import pickle
import requests
from bs4 import BeautifulSoup
allrecipes = pickle.load(open('allrecipes.p','rb'))
```

```python
recipes = list(set(allrecipes))
len(set(allrecipes))
```

# Getting Full HTML Data From Recipe Links in rawtextdata.p

```python
rawtextdata = []
for i in range(len(recipes)):
    r = requests.get(allrecipes[i])
    rawtextdata.append(r.text)
    if(i in [1000,2000,4000,6000,8000,10000]):
        pickle.dump(rawtextdata, open(f"rawtextdata{i//1000}k.p","wb"))
    if(i%10==0):
        print(i)
```

```python
pickle.dump(rawtextdata,open("rawtextdatafull.p","wb"))
```
